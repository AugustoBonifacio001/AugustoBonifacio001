```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 1000)
x = np.cos(4*t)
x1 =np.sin(4*t)

plt.figure("Cosseno ", figsize=(5, 4))
plt.plot(t,x)
plt.title("Gráfico do cosseno ")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.show()

plt.figure("Seno ",figsize=(5, 4))
plt.plot(t, x1)
plt.title("Gráfico do seno ")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude ")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Criar dados de exemplo
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Criar o gráfico principal
fig, ax = plt.subplots()

# Adicionar o gráfico principal
ax.plot(x, y, label='Gráfico Principal')

# Definir os limites do gráfico principal
ax.set_xlim(0, 6)
ax.set_ylim(0, 12)

# Adicionar o gráfico ampliado
axins = ax.inset_axes([0.6, 0.2, 0.3, 0.3])  # (x, y, largura, altura)

# Adicionar dados ao gráfico ampliado
axins.plot(x, y, label='Gráfico Ampliado')

# Definir os limites do gráfico ampliado
axins.set_xlim(2, 4)
axins.set_ylim(4, 8)

# Adicionar marcações de zoom
ax.indicate_inset_zoom(axins)

# Exibir o gráfico
plt.show()
```

```python
t = np.linspace(0, 2*np.pi, 1000)
x1 = np.cos(t)
x2 = np.sin(t)

plt.figure("Gráficos senoidais", figsize=(8, 4))
plt.subplots_adjust(
    left = 0.152,
    right = 1.52,



)

ax1 = plt.subplot(1, 2, 1)
plt.plot(t, x1)
ax1.set_title("Gráfico do cosseno")
ax1.set_xlabel("Eixo de tempo")
ax1.set_ylabel("Eixo de amplitude")

ax2 = plt.subplot(1, 2, 2)
plt.plot(t, x2)
ax2.set_title("Gráfico do seno")
ax2.set_xlabel("Eixo de tempo")
ax2.set_ylabel("Eixo de amplitude")

plt.show()
```
