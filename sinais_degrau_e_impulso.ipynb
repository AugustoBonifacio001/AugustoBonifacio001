```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = np.arange(0, 15, 1)
degrau= np.heaviside(u -4 , 1)

plt.figure(figsize=(8, 6))
plt.step(u, degrau)
plt.grid(True)
```

```python
u = np.arange(0, 15, 1)
degrau = np.heaviside(u-10, 0)
plt.plot(u, degrau)
plt.show()
```

```python
u = np.arange(0, 15, 1)
degrau = np.heaviside(u -4, 1) - np.heaviside(u -10, 0)
plt.plot(u, degrau)
plt.grid(True)
plt.show()
```

```python
# definindo o sinal impulso

u = np.arange(0, 15, 1)
impulso = np.where(u >= 4, 1, 0) #np.where(condição verdadeira, x(verdadeiro), y(falso))

plt.figure(figsize=(8, 6))
plt.stem(u, impulso)
plt.grid(True)
plt.show()
```
