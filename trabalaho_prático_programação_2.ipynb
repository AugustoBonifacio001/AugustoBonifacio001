```python
# Módulo para inicializar o tabuleiro
def inicializar_tabuleiro():
    """Inicializa o tabuleiro do jogo da velha e retorna uma lista representando o tabuleiro."""
    tabuleiro = []
    for i in range(3):
        linha = []
        for j in range(3):
            linha.append(" ")
        tabuleiro.append(linha)
    return tabuleiro

# Módulo para mostrar o tabuleiro
def mostrar_tabuleiro(tabuleiro):
    for linha in tabuleiro:
        print("|".join(linha))
        print("-" * 5)

# Módulo para obter a jogada de um jogador
def capturar_jogada(jogador):
    """Captura a jogada solicitando a linha e a coluna desejadas e retorna a tupla (linha, coluna)."""
    try:
        linha = int(input(f"Jogador {jogador}, informe a linha (0, 1 ou 2): "))
        coluna = int(input(f"Jogador {jogador}, informe a coluna (0, 1 ou 2): "))
        return linha, coluna
    except ValueError:
        print("Por favor, informe números válidos.")
        return capturar_jogada(jogador)

# Módulo para realizar uma jogada
def realizar_jogada(tabuleiro, jogador, linha, coluna):
    if tabuleiro[linha][coluna] == " ":
        tabuleiro[linha][coluna] = jogador
        return True
    else:
        print("Posição ocupada. Tente novamente.")
        return False

# Módulo principal para executar o jogo
def jogar_jogo_da_velha():
    """
    Função principal que executa o Jogo da Velha.
    """
    tabuleiro = inicializar_tabuleiro()
    jogador = "X"

    while True:
        mostrar_tabuleiro(tabuleiro)
        linha, coluna = capturar_jogada(jogador)

        if realizar_jogada(tabuleiro, jogador, linha, coluna):
            if verificar_vitoria(tabuleiro, jogador):
                mostrar_tabuleiro(tabuleiro)
                print(f"Jogador {jogador} venceu!")
                break
            elif verificar_empate(tabuleiro):
                mostrar_tabuleiro(tabuleiro)
                print("O jogo empatou!")
                break

            jogador = "O" if jogador == "X" else "X"

# Chama a função principal para iniciar o jogo
if __name__ == "__main__":
    jogar_jogo_da_velha()
```
