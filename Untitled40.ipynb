```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000
tx = 200
w = (2.0*np.pi)/tx

t = np.linspace(0, tx, n)

s1 = 2.0*np.cos(2.0*w*t)
s2 = 1.0*np.cos(30.0*w*t)

s = s1 + s2

freq =np.fft.fftfreq(n)
mascara = freq > 0
fft_calculo = np.fft.fft(s)
fft_abs = 2.0*np.abs(fft_calculo/n)

plt.figure(1)
plt.title("Sinal Original")
plt.plot(t, s)

plt.figure(2)
plt.title("Sinal da fft")
plt.plot(freq[mascara], fft_abs[mascara])

plt.show()
```

```python
n = 1000
fs =1000
t = np.linspace(0, 1, n , endpoint=False)

f1 =5
f2 = 20
sinal = 1.0*np.sin(2*np.pi*f1*t) + 0.5*np.sin(2*np.pi*f2*t)

freq = np.fft.fftfreq(n, 1/fs)
mascara = freq > 0
fft_calculo = np.fft.fft(sinal)
fft_abs = 2.0*np.abs(fft_calculo/n)

plt.figure(1)
plt.title("Sinal Original")
plt.plot(t, sinal)

plt.figure(2)
plt.title("Espectro de Frequência")
plt.plot(freq[mascara], fft_abs[mascara])

plt.show()
```

```python

n = 1000
fs = 1000
f1 = 5
f2 = 25
t = np.linspace(0, 1, n, endpoint= False)

signal = np.sin(2*np.pi*f1*t) + 2.0*np.sin(2*np.pi*f2*t)

freq = np.fft.fftfreq(n, 1/fs)
mascara = freq > 0
fft_calculo = np.fft.fft(signal)
fft_abs = 2*np.abs(fft_calculo/n)

plt.figure(1)
plt.plot(t, signal)

plt.figure(2)
plt.plot(freq[mascara], fft_abs[mascara])
plt.show()
```

```python
n = 1000
fs =1000
t = np.linspace(0 , 1, n, endpoint = False)

signal = np.sin(2*np.pi*f1*t) + 0.5*np.sin(2*np.pi*f2*t)

p =50
amostras_aleatorias = np.round((n-1)*np.random.rand(p)).astype(int)
sampled_signal = signal[amostras_aleatorias]


freq = np.fft.fftfreq(n, 1/fs)
mascara = freq > 0
fft_calculo = np.fft.fft(sampled_signal)
fft_abs = 2*np.abs(fft_calculo/p)
reconstructed_signal = np.fft.ifft(fft_calculo)
t_reconstructed= np.linspace(0, 1, len(reconstructed_signal), endpoint = False)

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, signal)
plt.title('Sinal Original')

plt.subplot(3, 1, 2)
plt.stem(t[amostras_aleatorias], sampled_signal, markerfmt='ro', basefmt ='k-')
plt.title('Amostragem Aleatória')

plt.subplot(3, 1, 3)
plt.plot(t_reconstructed, np.real(reconstructed_signal))
plt.title('Sinal Reconstruído')

plt.tight_layout()
plt.show()

```
