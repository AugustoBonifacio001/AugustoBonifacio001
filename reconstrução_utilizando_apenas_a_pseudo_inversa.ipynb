```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0.0, 0.1, 1000)
f = 5
x = np.sin(2*np.pi*f*t) + np.random.normal(0.0, 0.1, size=len(t))
plt.figure(figsize=(12,6))
plt.plot(t, x, label= "Sinal original")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.title("Sinal el√©trico original")
plt.legend()
plt.grid(True)
plt.show()

tx = 50
indices_amostrados = np.sort(np.random.choice(len(t), tx, replace= False))
y = x[indices_amostrados]
plt.figure(figsize=(12,6))
plt.plot(t, x, label= "Sinal original")
plt.scatter(t[indices_amostrados], y, color = 'red', label= "Pontos amostrados")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show()
print(len(t))

```

```python
Phi = np.zeros((tx, len(t)))
for i, idx in enumerate(indices_amostrados):
  Phi[i, idx] = 1



Phi_pseudo_inv = np.linalg.pinv(Phi)
print(Phi_pseudo_inv.shape)
print(y.shape)

sinal_reconstruido = Phi_pseudo_inv.dot(y)
plt.figure(figsize=(12, 6))
plt.plot(t, x, label='Sinal Original', alpha=0.7)
plt.plot(t, sinal_reconstruido, label='Sinal Reconstru√≠do', linestyle='--', linewidth=2)
plt.scatter(t[indices_amostrados], y, color='red', label='Pontos Amostrados')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Reconstru√ß√£o do Sinal El√©trico Utilizando Pseudo-Inversa')
plt.legend()
plt.grid(True)
plt.show()

# Calculando a taxa de erro (Erro Quadr√°tico M√©dio)
mse = np.mean((x - sinal_reconstruido)**2)
print(f'Taxa de Erro (MSE em porcentagem): {mse*100}')


```
