```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#criando dados fictícios
np.random.seed(42)
X= 2*np.random.rand(100,1)
Y = 4 + 3*X + np.random.randn(100,1)

#dividindo os dados em conjuntos de treino e de teste

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =0.2, random_state=42)

#criando um modelo de regressão linear:

model = LinearRegression()

#treinando o modelo:
model.fit(X_train, Y_train)

#fazendo previsões no conjunto de testes:
y_pred = model.predict(X_test)

#visualizando os resultados:

plt.scatter(X_test, Y_test, color = 'black', label='Real')
plt.plot(X_test, y_pred, color='blue', linewidth=3, label='Previsto')
plt.title("Regressão linear simples")
plt.xlabel("Variável Independente")
plt.ylabel("Variável Dependente")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

#criando dados fictícios
np.random.seed(42)
X= 2*np.random.rand(100,1)
Y = 4 + 3*X + np.random.randn(100,1)
print(X)
print(X.shape)
print(Y.shape)
print(Y)


#dividindo os dados em conjuntos de treino e de teste

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =0.2, random_state=42)

# Criando um modelo de regressão linear
model = LinearRegression()

# Treinando o modelo
model.fit(X_train, Y_train)

# Fazendo previsões no conjunto de teste
Y_pred = model.predict(X_test)

# Avaliando o desempenho do modelo
mse = mean_squared_error(Y_test, Y_pred)
mae = mean_absolute_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

# Exibindo as métricas
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'R-squared (R²): {r2:.2f}')

#visualizando os resultados:

plt.scatter(X_test, Y_test, color = 'black', label='Real')
plt.plot(X_test, y_pred, color='blue', linewidth=3, label='Previsto')
plt.title("Regressão linear simples")
plt.xlabel("Variável Independente")
plt.ylabel("Variável Dependente")
plt.legend()
plt.show()
```

```python
np.random.seed(42)
X = 2*np.random.rand(100,1)
Y= 4 + 3*X + np.random.randn(100,1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)
plt.scatter(X_test, Y_test, color='Black', label='Real')
plt.plot(X_test, Y_pred, color= 'blue', linewidth = 3, label="Previdto")
plt.title('Regressão Linear Simples')
plt.xlabel('Variável Independente')
plt.ylabel('Variável Dependente')
plt.legend()
plt.show()

# Avaliando o desempenho do modelo usando R²
r2 = r2_score(Y_test, Y_pred)  # Calcula o coeficiente de determinação R²
print(f'R-squared (R²): {r2:.2f}')  # Exibe o R²


```
