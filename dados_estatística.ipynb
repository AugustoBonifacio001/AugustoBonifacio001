```python
import pandas as pd
import numpy as np

# Dados transcritos da tabela para um DataFrame
data = {
    'Regiao': ['Brasil', 'Norte', 'Rondônia', 'Acre', 'Amazonas', 'Roraima', 'Pará', 'Amapá', 'Tocantins', 'Nordeste', 'Maranhão', 'Piauí', 'Ceará', 'Rio Grande do Norte', 'Paraíba', 'Pernambuco', 'Alagoas', 'Sergipe', 'Bahia', 'Sudeste', 'Minas Gerais', 'Espírito Santo', 'Rio de Janeiro', 'São Paulo', 'Sul', 'Paraná', 'Santa Catarina', 'Rio Grande do Sul', 'Centro-Oeste', 'Mato Grosso do Sul', 'Mato Grosso', 'Goiás', 'Distrito Federal'],
    'Branca_Total': [43.3, 19.5, 30.4, 18.1, 16.1, 20.4, 18.8, 18.1, 20.7, 25.1, 17.7, 22.5, 27.9, 39.3, 32.6, 31.3, 26.0, 24.8, 17.8, 51.0, 40.7, 37.8, 45.2, 59.1, 73.6, 63.9, 79.0, 79.9, 36.3, 43.5, 30.0, 36.3, 37.0],
    'Preta_ou_Parda_Total': [55.7, 78.9, 68.9, 80.4, 80.3, 76.6, 80.4, 81.3, 77.2, 74.1, 81.3, 77.5, 71.2, 60.4, 66.8, 68.0, 72.9, 74.9, 81.2, 47.8, 59.0, 61.6, 54.1, 39.1, 25.8, 35.1, 20.4, 19.9, 62.7, 54.9, 68.7, 62.9, 62.1],
    'Branca_10%_menores': [25.4, 13.9, 20.6, 13.4, 8.2, 15.9, 15.4, 13.3, 11.0, 18.6, 10.8, 18.6, 17.1, 32.9, 28.0, 22.9, 21.0, 16.4, 15.4, 36.3, 28.8, 26.9, 29.6, 46.1, 61.3, 55.1, 69.1, 66.6, 27.9, 28.3, 24.7, 31.0, 21.3],
    'Preta_ou_Parda_10%_menores': [73.6, 83.5, 78.6, 84.0, 85.3, 81.0, 83.5, 86.3, 82.1, 80.7, 87.8, 81.4, 82.5, 66.7, 72.0, 76.4, 78.3, 83.2, 84.0, 62.9, 71.0, 72.5, 69.6, 52.8, 38.2, 44.1, 30.4, 33.2, 71.3, 70.2, 74.5, 68.4, 78.1],
    'Branca_10%_maiores': [70.7, 35.7, 43.2, 29.6, 31.7, 37.9, 34.9, 34.6, 37.6, 40.5, 31.5, 40.4, 46.4, 57.2, 52.3, 47.7, 41.2, 38.4, 27.9, 76.0, 62.9, 62.8, 73.3, 81.6, 89.2, 82.8, 91.5, 94.1, 57.2, 63.8, 45.8, 56.7, 63.7],
    'Preta_ou_Parda_10%_maiores': [27.6, 63.2, 56.4, 69.9, 65.9, 60.7, 64.3, 64.9, 61.1, 59.0, 67.8, 59.6, 53.2, 42.8, 47.0, 51.8, 57.1, 60.8, 71.5, 21.4, 37.0, 36.6, 25.6, 14.5, 9.8, 14.9, 8.0, 5.7, 41.5, 34.5, 53.2, 42.6, 35.2]
}

df = pd.DataFrame(data)

# Função para calcular estatísticas descritivas
def calculate_statistics(column):
    mean = np.mean(column)
    median = np.median(column)
    std_dev = np.std(column)
    coef_variation = std_dev / mean * 100
    return mean, median, std_dev, coef_variation

# Calculando as estatísticas para as colunas de interesse
for column in df.columns[1:]:
    mean, median, std_dev, coef_variation = calculate_statistics(df[column])
    print(f"Estatísticas para {column}:")
    print(f"Média: {mean:.2f}")
    print(f"Mediana: {median:.2f}")
    print(f"Desvio Padrão: {std_dev:.2f}")
    print(f"Coeficiente de Variação: {coef_variation:.2f}%")
    print()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
categorias = ['Total', '10% menores', '10% maiores']
branca_media = [35.87, 27.02, 53.96]
preta_ou_parda_media = [63.09, 71.76, 45.00]

x = np.arange(len(categorias))  # A localização das categorias
largura = 0.35  # A largura das barras

fig, ax = plt.subplots()
barras1 = ax.bar(x - largura/2, branca_media, largura, label='Branca')
barras2 = ax.bar(x + largura/2, preta_ou_parda_media, largura, label='Preta ou Parda')

# Adicionar alguns textos para os rótulos, título e personalizar o eixo x
ax.set_xlabel('Categoria')
ax.set_ylabel('Média (%)')
ax.set_title('Média dos Percentuais de População')
ax.set_xticks(x)
ax.set_xticklabels(categorias)
ax.legend()

fig.tight_layout()
plt.show()
```

Explicação: Este gráfico de barras mostra a média dos percentuais de população branca e preta ou parda em três categorias diferentes: a população total, os 10% com menores rendimentos e os 10% com maiores rendimentos. Ele ilustra que a média de população branca é geralmente menor na categoria de menores rendimentos e maior na categoria de maiores rendimentos, enquanto a população preta ou parda apresenta a tendência oposta.
